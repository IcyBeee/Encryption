```python
from collections import Counter

# Takes a string or list of items and counts the frequencies of those items
# data: The list or string to analyse
# max_values: The maximum number of values to display (set to None for no limit)
# no_columns: The amount of columns to use in the output
def frequency_analysis(data, max_values=30, no_columns=5):
    frequencies = Counter()
    for item in data:
        frequencies[item] += 1
    
    total = sum(frequencies.values())
    column = 1
    for item, frequency in frequencies.most_common(max_values):
        print(f"{item}: {frequency:2} ({frequency / total:.2%})", end=" " if column % no_columns else "\n")
        column += 1
    print("\n-----")
```

```python
# Rail Fence decoder - developed for 2021 Mission 5a

def decrypt_rail_fence(message, rails):
    no_blocks = 2*(rails-1)
    message_length = len(message)
    block_size = message_length // no_blocks
    remainder = message_length % no_blocks
    block_sizes = list()
    for rail in range(rails):
        size = block_size
        if rail not in (0, rails-1):
            size *= 2
        block_sizes.append(size)

    for n in range(remainder):
        if n < rails:
            block_sizes[n] += 1
        else:
            block_sizes[rails-(n+2)] += 1

    blocks = list()

    start = 0
    for rail in range(rails):
        end = start + block_sizes[rail]
        block = message[start:end]
        if rail in (0, rails-1):
            block = "".join([letter+" " for letter in block])
        blocks.append(block)
        start = end

    text = ""
    for pos in range(0, max(block_sizes), 2):
        for block in range(rails):
            try:
                text += blocks[block][pos]
            except IndexError:
                pass
        for block in reversed(range(rails)):
            try:
                text += blocks[block][pos+1]
            except IndexError:
                pass

    print(text.replace(" ", ""))
```
