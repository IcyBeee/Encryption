These frequencies look remarkably similar to plain english.  Maybe columnar transposition is all I need to decode this?


```python
# Let's try to write a columnar transposition decoder

def columnar_transposition(text, key):
    text = text.replace(" ", "")
    columns = dict()
    column_length = len(text) // len(key)
    count = 0
    for column in sorted(key):
        start_pos = count*column_length
        columns[column] = text[start_pos:start_pos+column_length]
        count += 1
    for row in range(column_length):
        for char in key:
            print(columns[char][row], end="")

# Test it with the example on https://en.wikipedia.org/wiki/Transposition_cipher#Columnar_transposition
columnar_transposition("EVLNE ACDTK ESEAQ ROFOJ DEECU WIREE", "ZEBRAS")
```
