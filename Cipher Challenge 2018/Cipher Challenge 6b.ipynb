```python
from collections import Counter

# Takes a string or list of items and counts the frequencies of those items
# data: The list or string to analyse
# max_values: The maximum number of values to display (set to None for no limit)
# no_columns: The amount of columns to use in the output
def frequency_analysis(data, max_values=30, no_columns=5):
    frequencies = Counter()
    for item in data:
        frequencies[item] += 1
    
    total = sum(frequencies.values())
    column = 1
    for item, frequency in frequencies.most_common(max_values):
        print(f"{item}: {frequency:2} ({frequency / total:.2%})", end=" " if column % no_columns else "\n")
        column += 1
    print("\n-----")

frequency_analysis(squished)
```

```python
# The frequency looks like English, so this is a transposition cipher.
# How long is it, and what number of columns would fit easily?

length = len(squished)
print("Length = ", length)
for n in range(2, length):
    if length % n == 0:
        print(n, end=" ")
```

```python
# We have multiples of three and seven in there
# Looking at the text, I can see an anagram of "infiltrating the" at the beginning.
# Try organising into columns of seven, setting out "infiltr" in the first column

for pos in range(0, length, 7):
    print("".join(squished[pos+n] for n in (1, 0, 4, 3, 2, 6, 5)))
```
